## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Dataset Duplicate Value Count
Data.duplicated().sum()
````

````python
# Dataset Duplicate Value Count
Review.duplicated().sum()
````

````python
Data.drop_duplicates(inplace=True) # Removing duplicates from Data
Review.drop_duplicates(inplace=True) # Removing duplicates from Review
````

````python
# Dataset Duplicate Value Count
Data.duplicated().sum()
````

````python
# Dataset Duplicate Value Count
Review.duplicated().sum()
````

````python
# Missing values in Play Store dataset
Data.isnull().sum()
````

````python
# Missing values in User Reviews dataset
Review.isnull().sum()
````

````python
# Missing value percentage
(Data.isnull().sum() / len(Data)) * 100
````

````python
# Missing value percentage
(Review.isnull().sum() / len(Review)) * 100
````

````python
# Fill missing values

Data['Rating'] = Data['Rating'].fillna(Data['Rating'].median())

Data['Type'] = Data['Type'].fillna(Data['Type'].mode()[0])

Data['Content Rating'] = Data['Content Rating'].fillna(Data['Content Rating'].mode()[0])

Data['Current Ver'] = Data['Current Ver'].fillna(Data['Current Ver'].mode()[0])

Data['Android Ver'] = Data['Android Ver'].fillna(Data['Android Ver'].mode()[0])
````

````python
(Data.isnull().sum() / len(Data)) * 100   #check the percentage of null  values after fillna
````

````python
Review.dropna(inplace=True)  # Removing  missing values
````

````python
(Review.isnull().sum() / len(Review)) * 100 # checking missing values percentage after cleaning
````

````python
# Check data types of all columns in Play Store dataset
Data.dtypes
````

````python
# Check data types of all columns in User Reviews dataset
Review.dtypes
````

````python
# Remove invalid row(s) from Reviews column
Data = Data[Data['Reviews'].str.isnumeric()]

# Convert Reviews column to integer
Data['Reviews'] = Data['Reviews'].astype(int)
````

````python
# Remove ',' and '+' from Installs column
Data['Installs'] = Data['Installs'].str.replace(',', '', regex=False)
Data['Installs'] = Data['Installs'].str.replace('+', '', regex=False)

# Convert Installs column to integer
Data['Installs'] = Data['Installs'].astype(int)
````

````python
# Remove '$' symbol from Price column
Data['Price'] = Data['Price'].str.replace('$', '', regex=False)

# Convert Price column to float
Data['Price'] = Data['Price'].astype(float)
````

````python
# Replace 'Varies with device' with NaN
Data['Size'] = Data['Size'].replace('Varies with device', np.nan)

# Convert KB values to MB
Data['Size'] = Data['Size'].str.replace('k', '', regex=False)
Data['Size'] = Data['Size'].apply(lambda x: float(x)/1024 if pd.notnull(x) and 'M' not in str(x) else x)

# Remove 'M' from MB values
Data['Size'] = Data['Size'].astype(str).str.replace('M', '', regex=False)

# Convert Size column to float
Data['Size'] = pd.to_numeric(Data['Size'], errors='coerce')
````

````python
# Convert Last Updated column to datetime format
Data['Last Updated'] = pd.to_datetime(Data['Last Updated'])
````

````python
# Check updated data types
Data.dtypes
````
